## Введение
<br>Имеется мобильное приложение "Ненужные вещи". Приложение помогает людям продать ненужные вещи, просто разместив их на доске объявления. </br>
<br>Задача максимум - проверить имеющиеся гипотезы, а именно те, которые могли бы дать ответ на вопрос какой из этих двух групп показывают разное поведение - по-разному доходят к просмотру контактов. При этом такие метрики как конверсия и прочие метрики воздействии на платформе, а также анализ этих показателей дадут нам ответ не только о статистической значимости имеющейся разницы, но и плюс - в чью пользу.</br> 
<br>

5. На оснований анализа построить гипотезу, и подвергнуть ее проверке.

## Описание данных
Датасет **mobile_dataset.csv** содержит колонки:<br/>
`event.time` — время совершения<br/>
`event.name` — название события<br/>
`user.id` — идентификатор пользователя<br/>

Датасет **mobile_sources.csv** содержит колонки:<br/>
`userId` — идентификатор пользователя<br/>
`source` — источник, с которого пользователь установил приложение<br/>

***Расшифровки событий:***<br/>
`advert_open` — открытие карточки объявления<br/>
`photos_show` — просмотр фотографий в объявлении<br/>
`tips_show` — пользователь увидел рекомендованные объявления<br/>
`tips_click` — пользователь кликнул по рекомендованному объявлению<br/>
`contacts_show` и `show_contacts` — пользователь нажал на кнопку "посмотреть номер телефона" на карточке объявления<br/>
`contacts_call` — пользователь позвонил по номеру телефона на карточке объявления<br/>
`map` — пользователь открыл карту размещенных объявлений<br/>
`search_1` — `search_7` — разные события, связанные с поиском по сайту<br/>
`favorites_add` — добавление объявления в избранное

## Проверка гипотез
И так у нас две гипотезы:<br>
A. $H_{01}$: У группы, установивших приложение по ссылке из Yandex и группы, установивших приложение по ссылке из Google **конверсии в просмотры контактов** не имеют статистически значимую разницу.<br>
B. $H_{02}$: У группы, установивших приложение по ссылке из Yandex и группы, установивших приложение по ссылке из Google разные **конверсии в "добавить в избранное"** контактов не имеют статистически значимую разницу.<br>

```python
df.head(1)
```

```python
df_copy = df
df_copy.loc[df_copy['event_name']=='show_contacts', 'event_name']='contacts_show'
conversion = df_copy.pivot_table(index='event_name', columns='source', values='user_id', aggfunc=pd.Series.nunique)
#users_number=users_number.reset_index()
conversion.columns=['google', 'other', 'yandex']
conversion.reset_index(inplace=True)
display(conversion)
#.iloc[1]
user_number = df_copy.groupby('source')['user_id'].agg('count').reset_index()
user_number
```

"Обернем" наш **Z-тест** в функцию.

```python
def z_test(conversion_1, conversion_2, leads_1, leads_2):
    alpha=.05
    # определим все операции с переменными
    p1 = conversion_1/leads_1
    p2 = conversion_2/leads_2

    p_combined = (conversion_1 + conversion_2) / (leads_1 + leads_2)
    difference = p1 - p2 
    # определим уравнение z-теста
    z_value = difference / mth.sqrt(p_combined * (1 - p_combined) * (1/leads_1 + 1/leads_2))
    # определяем функцию для нормального распределения
    distr=st.norm(0, 1)
    # расчет p-value
    p_value = (1-distr.cdf(abs(z_value)))*2
    # вывести p-value
    print('p-значение: ', p_value)
    # вывести результат (условно)
    if (p_value<alpha):
        print("Отвергаем нулевую гипотезу: между долями есть значимая разница")
    else:
        print("Не получилось отвергнуть нулевую гипотезу, нет оснований считать доли разными")
```

### **$H_{01}$**:"У группы, установивших приложение по ссылке из Yandex и группы, установивших приложение по ссылке из Google конверсии в просмотры контактов не имеют статистически значимую разницу"
